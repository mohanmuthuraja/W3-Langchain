#Vector store

```python
pip install langchain-community faiss-cpu
```

```python
from langchain_classic.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
```

```python
text_source = [
    "Python is the best language for AI",
    "This course is for full beginners",
    "This will teach you all about Gen AI"
    
]

embedded_text = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

```

```python
print(embedded_text)
```

```python
faiss_index = FAISS.from_texts(text_source,embedded_text)
```

```python
faiss_index.save_local("my_first_vectorstore")
```

# Retriever.

```python
retriever = faiss_index.as_retriever()
```

```python
query = "python?"

answer = retriever.invoke(query)

print(answer)
```

```python
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python
text_source_croma = [
    "Python is the best language for AI",
    "This course is for full beginners",
    "This will teach you all about Gen AI"
    
]
```

```python
embedded_text_croma = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

```

```python

vector_text = Chroma.from_texts(
    text_source_croma,
    embedded_text_croma,
    collection_name="my_chroma_collection",
    persist_directory="my_chroma_vectorstore"
)
```

```python
vector_text.persist()
```
